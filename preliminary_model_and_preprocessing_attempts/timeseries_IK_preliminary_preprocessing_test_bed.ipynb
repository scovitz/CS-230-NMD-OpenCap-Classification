```python
# funcitons for reading .mot and .trc files

def read_trc(fpath):
    # read metadata in file header
    df_meta = pd.read_csv(fpath, delimiter='\t', header=0, skiprows=1, nrows=1)
    meta = df_meta.iloc[0].to_dict()
    fps = meta['DataRate']

    # read marker location names
    markers_df = pd.read_csv(fpath, delimiter='\t', header=None, skiprows=3, nrows=1)
    markers = markers_df.iloc[0].dropna().to_numpy()[2:]

    # read marker XYZ locations
    df = pd.read_csv(fpath, delimiter='\t', header=0, skiprows=4)
    df.rename(columns=dict(zip(df.columns[:2], ('n', 't'))), inplace=True)
    df.dropna(how='all', axis=1, inplace=True)

    N = df.shape[0]
    M = len(markers)
    xyz = df.iloc[:,2:].to_numpy().reshape((N, M, 3))
    xyz[:,:,[0,1,2]] = xyz[:,:,[2,1,0]]

    return fps, markers, xyz


def read_mot(fpath):
    with open(fpath, 'r') as f:
        # count = 0
        line = f.readline().strip()
        # while line and line.strip() != 'endheader':
        while line.lower() != 'endheader':
            line = f.readline().strip()
            # count += 1

        # df = pd.read_csv(f, delimiter='\t', header=0, skiprows=count-3)
        df = pd.read_csv(f, delimiter='\t', header=0)

    return df
```

```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from pathlib import Path

def load_data_for_classification(root_dir, window_size=100, step=10, normalize=True, sequence_length=900):
    """
    Loads .mot files for each participant-trial pair, aggregates the data by trial,
    and prepares it for LSTM input with one output classification per trial.
    
    Parameters:
        root_dir (str): Root directory containing participant folders.
        window_size (int): Number of time steps in each window.
        step (int): Step size between windows.
        normalize (bool): Whether to normalize each feature across each file.
        sequence_length (int): Fixed length for each sequence (pad or truncate).
    
    Returns:
        X (list): List of 3D numpy arrays, each representing data for one participant-trial.
        y (list): List of classification labels (0 or 1) for each participant-trial.
    """
    # Initialize lists for data and labels
    X, y = [], []
    
    # Expected columns (dropping 'time' as it's not needed for modeling)
    expected_columns = [
        'pelvis_tilt', 'pelvis_list', 'pelvis_rotation', 'pelvis_tx',
        'pelvis_ty', 'pelvis_tz', 'hip_flexion_r', 'hip_adduction_r',
        'hip_rotation_r', 'knee_angle_r', 'ankle_angle_r',
        'subtalar_angle_r', 'mtp_angle_r', 'hip_flexion_l', 'hip_adduction_l',
        'hip_rotation_l', 'knee_angle_l', 'ankle_angle_l',
        'subtalar_angle_l', 'mtp_angle_l', 'lumbar_extension', 'lumbar_bending',
        'lumbar_rotation', 'arm_flex_r', 'arm_add_r', 'arm_rot_r',
        'elbow_flex_r', 'pro_sup_r', 'arm_flex_l', 'arm_add_l', 'arm_rot_l',
        'elbow_flex_l', 'pro_sup_l'
    ]
    
    # Dictionary to store activities for each participant-trial
    participant_trial_data = {}

    for i in range(129):
        num = str(i+1).zfill(3)
        
        for path in Path(root_dir).rglob(f"P{num}*/**/walk.mot"): # START WITH JUST WALK TRIALS
            trial = str(path).split('/')[-3]
            participant_id = str(path).split('/')[-4]
            activity = str(path).split('/')[-2]

            # Key by participant-trial to aggregate activities
            participant_trial_key = f"{participant_id}_{trial}"
            
            # Initialize list for activities within the trial if not already present
            if participant_trial_key not in participant_trial_data:
                participant_trial_data[participant_trial_key] = []
            
            # Load the data
            data = read_mot(str(path))

            # Ensure the file has the expected columns
            if not all(col in data.columns for col in expected_columns + ['time']):
                raise ValueError(f"File {path} does not contain the expected columns.")
            
            # Drop the 'time' column and keep only relevant features
            data = data[expected_columns]

            # Normalize features if specified
            if normalize:
                scaler = StandardScaler()
                data = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
            
            # Append the processed activity data to the list for this trial
            participant_trial_data[participant_trial_key].append(data)

    # Aggregate data across activities for each participant-trial and apply windowing
    for participant_trial_key, activities in participant_trial_data.items():
        # Concatenate all activity data within the trial along the time axis
        trial_data = pd.concat(activities, ignore_index=True)
        
        # Pad or truncate trial_data to sequence_length
        if len(trial_data) < sequence_length:
            # Pad with zeros if sequence is shorter than the desired length
            padding = pd.DataFrame(np.zeros((sequence_length - len(trial_data), len(expected_columns))), columns=expected_columns)
            trial_data = pd.concat([trial_data, padding], ignore_index=True)
        elif len(trial_data) > sequence_length:
            # Truncate if sequence is longer than the desired length
            trial_data = trial_data.iloc[:sequence_length]
        
        # Create sliding windows for the aggregated trial data
        num_windows = (len(trial_data) - window_size) // step + 1
        windows = np.zeros((num_windows, window_size, len(expected_columns)))

        for i in range(num_windows):
            start = i * step
            end = start + window_size
            windows[i] = trial_data.iloc[start:end].values
        
        # Append the windows to X
        X.append(windows)

        # Placeholder for label (assuming you have a separate function or method to get labels)
        participant_id, trial = participant_trial_key.split('_')
        label = get_label_for_participant_trial(participant_id, trial)
        y.append(label)

    # Convert X and y to numpy arrays
    X = np.array(X, dtype=object)  # Using dtype=object due to variable number of windows per trial
    y = np.array(y)
    
    return X, y

def get_label_for_participant_trial(pid, trial):
    """
    Fetches the label for a given participant and trial based on class_info.csv.
    
    Parameters:
        participant_id (str): The ID of the participant.
        trial (str): The trial name or ID.
    
    Returns:
        int: Classification label (0 or 1).
    """
    class_df = pd.read_csv('/Users/scovitz/datadir/class_info.csv')
    label = class_df.loc[class_df['ID'] == pid, 'Class'].values[0]
    return label
```
